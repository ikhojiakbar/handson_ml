# 1. Training and Visualizing a Decision Tree

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = datasets.load_iris()
X = iris['data'][:, 2:] #petal length and with
y = iris['target']
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

# 2. Making Predictions
* Gini Impurity = 0 is the ideal value for root node
* Each depth sets only one features threshold value only once, the next child nodes set another feature threshold only once
* sklearn uses CART algorithm under the hood => binary trees only (2 child nodes as True/False)
* Alternative ID3 can produce more than 2 child nodes

## 2.1 Decision boundaries of each depth
* Dotted lines are used if we set `max_depth` = 3

# 3. Estimating Class Probabilities
* Both `predict()` and `predict_proba()` are supported

```python
tree_clf.predict([[5, 1.5]])
```

```python
iris['target_names'][tree_clf.predict([[5, 1.5]])]
```

```python
#90.7% = 49/54 based on training instances split
tree_clf.predict_proba([[5, 1.5]])
```
