## 1. Soft Margin Classification
* Hard Margin classification = instances should be linearly separable and don't be outliers (N. ans S. Korea)
* Soft Margin Classification = no strict punishing to outliers

### 1.1 Margin violations (C hyperparameter)
* If C is low, Margin violation is high = GENERALIZES WELL
* If C is high, Margin violation is low = STRICT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.svm import LinearSVC
```

```python
iris = datasets.load_iris()

X = iris['data'][:, (2, 3)] #petal length and width
y = (iris['target'] == 2).astype(np.float64)
```

* Cost Function = Hinge
* Activation and Optimization algorithms the same as Logistic one

```python
svm_clf = Pipeline([
    ('scaler', StandardScaler()),
    ('linear_svc', LinearSVC(C=1, loss='hinge'))
])

svm_clf.fit(X, y)
```

* No probability for output class

```python
#virginica
svm_clf.predict([[5.5, 1.7]])
```
