# 1. CART Algorithm
* Classification and Regression Tree
* The algorihm worksby splitting the training set into 2 subsets using a single feature K and threshold of K
* The algorithm seraches for the feature K and K-threshold that returns the minimum Gini Impurity
* And so on recursively, therefore we need stopping conditions for the recursion.
* It is a `greedy algorithm`
* Hyperparameters that STOP the recursion:
* 1. `max_depth`
* 2. `min_samples_split`
* 3. `min_samples_leaf`
* 4. `min_weight_fraction_leaf`
* 5. `max_leaf_nodes`

# 2. Computational Complexity
* Prediction: O(log(m))
* Training: O(n x mlog(m))

# 3. Gini Impurity or Entropy? 
* By default, Gini Impurity
* Reset by `criterion='entropy'`
* Gini Impurity is good for working with the most frequent class (imbalanced class)
* Entropy tries to produces more balanced trees

# 4. Regularization Hyperparameters
* DT are prone to overfitting => need regularization
* 1. `max_depth`, by default `None` => the most important hyperparameter
* 2. `min_samples_split`
* 3. `min_samples_leaf`
* 4. `min_weight_fraction_leaf`
* 5. `max_leaf_nodes`
* 6. `max_features`
