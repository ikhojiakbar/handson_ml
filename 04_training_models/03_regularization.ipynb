## 1.1 Ridge Regression (L2)
* Added to cost function only when the model is trained
* `alpha` is the hyperparameter if alpha = 0 = just LR, if it is large => reducing heavily
* StandardScaler is mandatory for regularization technique
* L2 is handy with Polynomial models => reduces the degree of the model
* Can be modeled either SVD or GD

### 1.1.1 SVD

```python
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=1, solver='cholesky')
ridge_reg.fit(X, y)
```

### 1.1.2 SGD

```python
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(penalty='l2')
sgd_reg.fit(X, y)
```

## 1.2 Lasso Regression (L1)
