# SVM Regression
* Objective = fit as many as possible instance ON the street (bounds= support vectors)
* The width of the street is controlled by Epsilon hyperparameter

## 1. LinearSVR

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=100, noise=0.15)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVR
```

```python
svm_reg = Pipeline([
    ('scaler', StandardScaler()),
    ('svm_reg', LinearSVR(epsilon=1.5))
])
```

```python
svm_reg.fit(X, y)
```

## 2. Polynomial SVR
* Poly kernel method
* Model itself is polynomial

```python
from sklearn.svm import SVR
```

```python
poly_svm_reg = Pipeline([
    ('scaler', StandardScaler()),
    ('poly_svm_reg', SVR(kernel='poly', degree=2, C=100, epsilon=0.1))
])
```

```python
poly_svm_reg.fit(X, y)
```

SVMs can be used to outlier detection problem
